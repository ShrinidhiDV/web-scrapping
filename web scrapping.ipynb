```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# List of category URLs (example)
category_urls = [
    'https://www.myntra.com/men-tshirts',
    'https://www.myntra.com/men-shirts',
    'https://www.myntra.com/men-jeans',
    # Add more categories as needed
]

# Set up the Chrome WebDriver using ChromeDriverManager
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Function to scrape product details from a page
def scrape_products(soup):
    products = soup.find_all('li', class_='product-base')
    for product in products:
        try:
            # Product name
            name = product.find('h4', class_='product-product').text
            
            # Product brand
            brand = product.find('h3', class_='product-brand').text
            
            # Product price
            price = product.find('span', class_='product-discountedPrice').text if product.find('span', class_='product-discountedPrice') else product.find('span', class_='product-strike').text
            
            # Print details
            print(f"Brand: {brand}")
            print(f"Name: {name}")
            print(f"Price: {price}")
            print('-' * 40)
        except AttributeError:
            # Handle missing elements gracefully
            print("An element was missing from this product's details.")

# Function to handle pagination
def scrape_category(category_url):
    driver.get(category_url)
    driver.implicitly_wait(10)

    while True:
        # Get the page source and parse it with BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Scrape product details from the current page
        scrape_products(soup)
        
        # Find the "Next" button to go to the next page
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, 'a.pagination-next')
            next_button.click()
            time.sleep(3)  # Wait for the next page to load
        except:
            break  # No more pages

# Iterate over each category URL
for url in category_urls:
    print(f"Scraping category: {url}")
    scrape_category(url)

# Close the WebDriver
driver.quit()
```
